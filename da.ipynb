```python

import pandas as pd
import datetime


file_type=".csv"
base_link="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"
```

```python
def dailyActive(date,country):
    dataframe=pd.read_csv(base_link+date+file_type)
    da=dataframe.loc[dataframe["Country_Region"]==country]["Active"].values[0]
    return da
```

```python
def activeRange(country,numOfDaysInPast):

    """Gives number of active cases from today to past n number of days, skips until data is available"""
    dateList=[]
    for i in range(0,numOfDaysInPast):
        pastdate=datetime.date.today()-datetime.timedelta(days=i)
        past=pastdate.strftime('%m-%d-%Y')
        dateList.insert(0,past)
    activeList=[]
    for day in dateList:
        try:
            activeList.append(dailyActive(day,country))
        except:
            pass
    
    return activeList
```

```python
def countryActiveRange(countryList, numOfDaysInPast):
    activeDict={}
    for country in countryList:
        activeDict[country]=activeRange(country,numOfDaysInPast)
    return activeDict
```

```python
countries=["Ireland","Poland","Serbia","India"]

#note some countries have provincial data as well, in this case we have to pass province,country to get data, check github keys and type in exact text, I'll find a way to make this easier in the future
```

```python
ab=countryActiveRange(countries,7)
```
